As you can see there is a space in between the column labels. Hence, in order to make my visualization easy I will now need to clean that space. After the cleaning is done my column labels will look like from ``Grade Level`` to ``Grade_level``. 

```python
sns.set(rc={"figure.figsize":(20,20)})
ax = sns.scatterplot(data = df, 
                y = 'Teacher', 
                x = 'Grade Level',
                x_bins= 10, y_bins= 10,
                legend='auto', n_boot=1000, alpha =1, x_jitter= None,
                ci= None
                )
#sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```
